```python
import pandas as pd 
import numpy as np
```

```python
""" 
Loading data from directory 
"""


df = pd.read_csv('D:\\1-Data Science\\2-Dataset\\wbcd.csv')
```

```python
""" 
Checking number of colummns in dataset 
"""

len(df.columns)
```

```python
""" 
Checking five numbers summary for detailes
"""

df.describe()
```

```python
""" 
1. Checking columns null and non null values
2. Checking Data types 
"""

df.info()
```

```python
df
```

```python
"""
 Changing the value B in df['diagnosis] to Beniegn
"""

df['diagnosis'] = np.where(df['diagnosis']=='B','Beniegn', df['diagnosis'])
```

```python
"""
 Changing the value M in df['diagnosis'] to Malignant
"""

df['diagnosis'] = np.where(df['diagnosis']=='M','Malignant', df['diagnosis'])
```

```python
df
```

```python
df.columns
```

```python
df
```

```python
""" 
Seperating the features and target columns from data 
"""

df = df.iloc[:,1:32]
df
```

```python
""" 
Function for normalization of data due to variance 
"""

def norm_funt(i):
    x = (i-i.min())/(i.max()-i.min())
    return x
```

```python
""" 
Nomalized data and stored in variable df_n
"""
df_n =norm_funt(df.iloc[:,1:32])
df_n
```

```python
""" 
Converting this df_n to array 
"""

X = np.array(df_n)
X
```

```python
""" 
Converting target variable to array
"""

y = np.array(df['diagnosis'])
y
```

Splitting the given data into 80-20 for the training and testing purpose :-

```python
from sklearn.model_selection import train_test_split
```

```python
""" 
Splitting data using train_test_split in 80-20 ratio for training and testing of model
"""

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

Selecting the KNN model for prediction purpose :-

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
""" 
Creating model (K nearest Neighbour model) with number of neighbours assigned 21
"""

model = KNeighborsClassifier(n_neighbors=21)
```

```python
""" 
Fitting training data inside model
"""

model.fit(X_train, y_train)
```

```python
""" 
Predicting on testing data using trained model
"""

pred = model.predict(X_test)
pred
```

Calculating The accuracy of the Model :-

```python
from sklearn.metrics import accuracy_score
```

```python
"""  
Finding accuracy of model
"""

print(accuracy_score(pred, y_test))
```

```python
"""
Confusion matrix for model
"""
pd.crosstab(pred, y_test)
```

```python
acc = []
```

```python
for i in range(3,50,2):
    neigh = KNeighborsClassifier(n_neighbors=i)
    neigh.fit(X_train, y_train)
    train_acc = np.mean(neigh.predict(X_train)==y_train)
    test_acc = np.mean(neigh.predict(X_test)== y_test)
    acc.append([train_acc, test_acc])
```

```python
acc
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(np.arange(3,50,2), [i[0] for i in acc], "ro-" )
plt.plot(np.arange(3,50,2),[i[1] for i in acc], "bo-")
```

```python
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train, y_train)
pred = knn.predict(X_test)
accuracy_score(pred, y_test)
pd.crosstab(pred, y_test)
```
